```python
import matplotlib.pyplot as plt
import pandas as pd
numpy as np
data={"Bill Source":["Electricity Bill","Mobile Recharge","Dish Recharge","Water bill","Internet Bill","Garbage picking Bill"],
     "Bill Amount":[2500,350,355,500,834,200]}
data=pd.DataFrame(data)
data
```

```python
data.to_csv("bill_report.csv",index=False)
```

```python
a=data['Bill Source']
b=data['Bill Amount']
c=list(a)
d=list(b)
plt.title('Bills data')
plt.pie(d,labels=c,autopct='%.2f%%',radius=1,shadow=True,explode=[0,0,0,0.2,0,0])
plt.show()
```
