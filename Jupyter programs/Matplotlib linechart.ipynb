```python
import matplotlib.pyplot as plt
days=[2,5,6,7,8,10,11]
b_max=[45,50,55,80,55,34,80]
b_min=[45,37,42,50,53,39,30]
c_ave=[35,38,42,45,45,67,75]
plt.plot(days,b_max,marker='',linestyle='--',markersize='5',label='Max temp')
plt.plot(days,b_min,marker='',linestyle='--',markersize='5',label='Min temp')
plt.plot(days,c_ave,marker='',linestyle='--',markersize='5',label='Ave temp')
plt.legend(loc="best",shadow=True)
```

```python
import matplotlib.pyplot as plt
days=['Mon','Tue','Wed','Thu','fri','Sat','Sun']
b_max=[50,51,50,49,49,50,51]
plt.plot(days,b_max,label='temp')
plt.xlabel('Days')
plt.ylabel('Tmep')
plt.title('Weather data table')
plt.legend()
```

```python
import matplotlib.pyplot as plt
days=[2,5,6,7,8,10,11]
b_max=[45,50,55,80,55,34,80]
b_min=[45,37,42,50,53,39,30]
c_ave=[35,38,42,45,45,67,75]
plt.plot(days,b_max,marker='.',linestyle='solid',markersize='5',label='Max temp')
plt.plot(days,b_min,marker='.',linestyle='solid',markersize='5',label='Min temp')
plt.plot(days,c_ave,marker='.',linestyle='solid',markersize='5',label='Ave temp')
plt.legend(loc="best",shadow=True)
plt.grid()
```
