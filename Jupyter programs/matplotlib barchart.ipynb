```python
##horizontal Barchart
import matplotlib.pyplot as plt
import numpy as np
company=["Car dekho","Ola","Cred","Flipkart","Delloite"]
revenue=[100,300,350,450,600]
str_c=np.arange(len(company))
plt.yticks(str_c,company)
plt.xlabel("Revenue in Crore")
plt.ylabel("Company names")
plt.title("Revenue bar chart")
plt.barh(str_c+0.2,revenue,height=0.4,label="Revenue")
plt.barh(str_c-0.2,profit,height=0.4,label="Profit")
plt.legend()
```

```python
##vertical Barchart

import matplotlib.pyplot as plt
import numpy as np
company=["Car dekho","Ola","Cred","Flipkart","Delloite"]
revenue=[100,300,350,450,600]
str_c=np.arange(len(company))
plt.xticks(str_c,company)
plt.ylabel("Revenue in Crore")
plt.xlabel("Company names")
plt.title("Revenue bar chart")
plt.bar(str_c-0.2,revenue,width=0.4,label="Revenue",color="orange")
plt.bar(str_c+0.2,profit,width=0.4,label="Profit",color="Green")
plt.legend()
```
